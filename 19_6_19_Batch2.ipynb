```python
### Deduct 10 from current Sales when Ship Mode is First Class
## apply()
# map over a sequence(v1,v2,v3,...vn)===> (f(v1),f(v2),f(v3)....f(vn))
# table is a sequence of what?
# (f(col1),f(col2),f(col3)...)
# (f(r1),f(r2),....f(rn))

def normalize(x):
    if x['ShipMode']=='First Class':
        return x['Sales']-10
    else:
        return x['Sales']
```

```python
data['Sales_Normalized']=data.apply(normalize,axis=1)
```
