```python
import pandas as pd
path="/Users/gunnvantsaini/Store.csv"
data=pd.read_csv(path)
data.head(2)
```

```python
### Label transactions where the product category is Furniture and Sales >500 as 
# imp transactions
```

```python
def get_imp(x):
    if x.Category=='Furniture' and x.Sales>500:
        return "ImpCust"
    else:
        return "NotImp"
data.apply(get_imp,axis=1)
```
