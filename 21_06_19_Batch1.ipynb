```python
## How the avg profit was distributed across product categories?
# variable which groups the data ====> category
# variable which is summarised/agg ====> profit
# how the variable is agg =====> avg/mean
data.groupby('Category').agg({'Profit':'mean'})
```

```python
### We can add more than one grouping variable
## How the avg profit was distributed across product categories and regions?
data.groupby(['Category','Region']).agg({'Profit':'mean'}) ### Col:agg
```

```python
## Only agg one column in one way
## How the avg and max profit as well as avg and max discount  
## was distributed across product categories and regions?
## agg which cols =====> Discount, Profit
## How =====> mean and max
## grouping vars ====> Category+Region
```

```python
data.groupby(['Category',"Region"]).agg({'Profit':['mean','max'],
                                         'Discount':['mean','max']})
```

```python
def get_sales(x):
    if x['ShipMode']=='First Class':
        return x['Sales']-10
    else:
        return x['Sales']
```

```python
data['Sales_adj']=data.apply(get_sales,axis=1)
```
